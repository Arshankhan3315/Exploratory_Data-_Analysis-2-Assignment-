# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

# Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

# Answer-5-

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=flight_prices)
plt.title('Boxplot of Flight Prices')
plt.show()
```

```python
Q1 = flight_prices.quantile(0.25)
Q3 = flight_prices.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
potential_outliers = flight_prices[(flight_prices < lower_bound) | (flight_prices > upper_bound)]
```

```python
# Print potential outliers
print("Potential outliers:")
print(potential_outliers)
```

# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
size_rating_data = playstore_data[['Size', 'Rating']]
```

```python
size_rating_data['Size'].unique()
```

# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

# Answer-13-

```python
type_price_data = playstore_data[['Type', 'Price','Installs']]
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        type_price_data[cols]=type_price_data[cols].str.replace(item,'')
```

```python
type_price_data['Price']=type_price_data['Price'].replace('Everyone',np.nan)
```

```python
type_price_data['Price'].unique()
```

```python
type_price_data['Price']=type_price_data['Price'].astype('float')
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Type', y='Price', data=type_price_data, ci=None)
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

# Answer-14-

```python
# Assuming "App" is the column containing app names and "Installs" is the column containing the number of installs
app_installs_data = playstore_data[['App', 'Installs']]
```

```python
# Create a frequency table
app_frequency_table = app_installs_data.groupby('App')['Installs'].sum().reset_index()

# Sort the table in descending order of installs
app_frequencyfloatble = app_frequency_table.sort_values(by='Installs', ascending=False)

# Display the top 10 most popular apps
top_10_apps = app_frequency_table.head(10)
print("Top 10 Most Popular Apps:")
print(top_10_apps)
```
